## Stacking

```python
a = np.floor(10*rg.random((2,2),dtype=float))
b = np.floor(10*rg.random((2,2),dtype=float))
```

```python
a , b
```

```python
np.vstack((a,b)),np.hstack((a,b))
```

```python
from numpy import newaxis
```

```python
np.column_stack((a,b))
```

```python
a[:newaxis]
```

```python
np.argmax(a),np.min(a),a.argmin(),a.max(),np.sort(a)
```
